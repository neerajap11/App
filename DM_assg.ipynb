```
#Processing indicators
import csv
with open('Data.csv', 'r') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',')
    data = list(spamreader)
    data = np.asarray(data)
    dt = pd.DataFrame(data=data[1:-1556], columns=data[0])
    dt.drop(columns=['Country Code', 'Series Name'], axis=1, inplace=True)
    dt.columns = ['Country', 'Series', '2013', '2014', '2015', '2016','2017','2018']
    dt.loc[dt['2015'] == '..', '2015'] = {}
    dt.loc[dt['2016'] == '..', '2016'] = {}
    dt.loc[dt['2017'] == '..', '2017'] = {}
    dt['2015'] = dt['2015'].astype(float)
    dt['2016'] = dt['2016'].astype(float)
    dt['2017'] = dt['2017'].astype(float)
    dt_train= dt[['Country', 'Series', '2015', '2016']]
    dt_test=dt[['Country','Series','2017']]
```

```
from sklearn import linear_model
from sklearn.metrics import r2_score
```

```
dt_i=pd.read_csv("dt_train_filled.csv")
dt_ii=pd.read_csv("dt_test_filled.csv")

dt_i=dt_i.loc[:, ~dt_i.columns.str.contains('^Unnamed')]
dt_ii=dt_ii.loc[:, ~dt_ii.columns.str.contains('^Unnamed')]

dt_i.to_pickle('happiness_traindata.pkl')
dt_ii.to_pickle('happiness_testdata.pkl')
dp_i = pd.read_pickle('happiness_traindata.pkl')
dp_ii = pd.read_pickle('happiness_testdata.pkl')
```

```
years_train = [2015, 2016]
years_test = [2017]
train_features = dp_i.loc[dp_i['Year'].isin(years_train)]
test_features = dp_ii.loc[dp_ii['Year'].isin(years_test)]
train_features.reset_index(inplace=True, drop=True)
test_features.reset_index(inplace=True, drop=True)

df_train = train_features.drop(columns=['Country', 'Year', 'Happiness Score'], axis=1)
df_test = test_features.drop(columns=['Country', 'Year', 'Happiness.Score'], axis=1)
df_test
```

```
df_test1=df_test.fillna(df_train.mean())
df_test1
```

```
df_train
```

```
happiness_train = pd.DataFrame(train_features['Happiness Score'])
happiness_test = pd.DataFrame(test_features['Happiness.Score'])

lr = linear_model.LinearRegression()
lr.fit(df_train, happiness_train.values.ravel())
happiness_predictions = lr.predict(df_test1)
errors = abs(happiness_predictions - happiness_test.values.ravel())
accuracy = r2_score(happiness_test.values.ravel(), happiness_predictions)
print('R2 score: ', accuracy)
m = 100 * (errors / happiness_test.values.ravel())
accuracy = 100 - np.mean(m)
print('Accuracy: ', accuracy)
errors
```

```
from sklearn.ensemble import RandomForestRegressor

happiness_train = pd.DataFrame(train_features['Happiness Score'])
happiness_test = pd.DataFrame(test_features['Happiness.Score'])

#lr = linear_model.LinearRegression()
rf = RandomForestRegressor(n_estimators=5000)
rf.fit(df_train, happiness_train.values.ravel())
happiness_predictions = rf.predict(df_test1)
errors = abs(happiness_predictions - happiness_test.values.ravel())
accuracy = r2_score(happiness_test.values.ravel(), happiness_predictions)
print('R2 score: ', accuracy)
m = 100 * (errors / happiness_test.values.ravel())
accuracy = 100 - np.mean(m)
print('Accuracy: ', accuracy)
errors
```

```
happiness_train = pd.DataFrame(train_features['Happiness Score'])
happiness_test = pd.DataFrame(test_features['Happiness.Score'])


# import libraries
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error

# steps
steps = [('scaler', StandardScaler()),
         ('ridge', Ridge())]

# Create the pipeline: pipeline
pipeline = Pipeline(steps)

# Specify the hyperparameter space
parameters = {'ridge__alpha':np.logspace(-4, 0, 50)}

# Create the GridSearchCV object: cv
cv = GridSearchCV(pipeline, parameters, cv=3)

# Fit to the training set
cv.fit(happiness_train, happiness_train)

#predict on train set
#y_pred_train=cv.predict(happiness_train)

# Predict test set
y_pred_test=cv.predict(happiness_test)

# rmse on train set
#rmse = np.sqrt(mean_squared_error(happiness_train, y_pred_train))
#print("Root Mean Squared Error: {}".format(rmse))
errors = abs(y_pred_test - happiness_test.values.ravel())
accuracy = r2_score(happiness_test.values.ravel(), y_pred_test)
print('R2 score: ', accuracy)
m = 100 * (errors / happiness_test.values.ravel())
accuracy = 100 - np.mean(m)
print('Accuracy: ', accuracy)
errors
```
